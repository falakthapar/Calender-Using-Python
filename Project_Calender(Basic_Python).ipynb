### Step 00 : Declaring 2 lists:
* 1st contain  12 months of the year
* 2nd contain number of days in each month.

```python
month_names = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
month_numdays = [31,28,31,30,31,30,31,31,30,31,30,31]
```

###Step 01 : Creating a Function to check for Leap Year (Return 1 if Leap Year, 0 otherwise)

```python
def is_leap(year):
    if(year % 400 == 0):
      return 1
    elif ((year % 100 != 0) and  (year % 4 == 0)):
      return 1
    else:
      return 0
```

### Step 02 : Creating a Dictionary with Year name  and number of days in each year (considering Leap year)
* Getting the year as a input to print the calender .
* Making a dictionary which will contain key as years from 1600-2100 and values as no of days in that year, considering leap years.
* Making use of the above dictionary to get the another dictionary which contain year as a key and day of 1st Jan of that year as a value.
* Using the dictionary to fetch day of 1st January.
* Here I have reprsented each 'Day of the Week' with a number between 1-7
 * Mon:1, Tue:2, Wed:3, Thu:4, Fri:5, Sat:6, Sun:7

```python
cal_year = int(input('Enter the year of which you want to see the calender  - '))

Year_days_dict = {}

years = [i for i in range(1600,2100)]
No_of_days = []

for i in years:
  if is_leap(i) == 1:
    No_of_days.append(366)
  else:
    No_of_days.append(365)

year_1_DOW = 6

years_DOW = [year_1_DOW]
y = year_1_DOW

for i in No_of_days:
  x = i%7
  z = (x+y)%7
  if z == 0:
    z = 7
  years_DOW.append(z)
  y = z
years_DOW.pop()



years_DOW_dict = {}

for i,j in zip(years,years_DOW):
  years_DOW_dict[i] = j


Jan_1_DOW = years_DOW_dict[cal_year]
```

### Step 03 : Creating a Dictionary with Month Name and No of days in each month (considering Leap year)
* Made use of the Two Lists declared in Step 00
* For leap years number of days in Feb is changed to 29.

```python
mth_numdays_dict = {}

if is_leap(cal_year) == 1:
  month_numdays[1] = 29
mth_numdays_dict = {}
for i,j in zip(month_names,month_numdays):
  mth_numdays_dict[i] = [j]

mth_numdays_dict
```

###Step 04 : Finding which 'Day of the Week' each Month begins with and adding it to the Dictionary
* For this step, made use of the variable Jan_1_DOW declared in Step 02
* Represented each 'Day of the Week' with a number between 1-7
 * Mon:1, Tue:2, Wed:3, Thu:4, Fri:5, Sat:6, Sun:7

```python
months_DOW = [Jan_1_DOW]
y = Jan_1_DOW

for i in month_numdays:
  x = i%7
  z = (x+y)%7
  if z == 0:
    z = 7
  months_DOW.append(z)
  y = z
months_DOW.pop()

months_DOW

for i,j in zip(mth_numdays_dict.values(),months_DOW):
  i.append(j)


mth_numdays_dict
```

### Step 05 : Creating a list 'cal_mth' with Month dates as strings and leading values as blanks

* For dates < 10, added leading Zeroes to make the String length = 2.
* Added blank values of double spaces in the in the beginning of the list 'cal_mth'.
* The count of these blank values in the List will depend on DOW (Day of Week) of First Day of the month. E.g.
 * If First Day of the month falls on a Monday, then no blank value will appear
 * If First Day of the month falls on a Tuesday, then 1 blank value will appear
 * ... and so on.

```python
which_month = 'Jan'
mth = mth_numdays_dict[which_month]
cal_mth = []

for i in range(0,mth[1]-1):
  cal_mth = cal_mth + ['  ']
for i in range(1,mth[0]+1):
  x = str(i)
  if len(x) == 1:
    x = '0'+x
  cal_mth = cal_mth + [x]

print(cal_mth)

```

### Step 06 : Publishing calendar for a particular month
* Developed a code in this cell such that it published the Calendar view for any month.
* Here I have used month which is used in the last cell.

```python
print('='*20)
print(which_month)
print('='*20)
print('Mo','Tu','We','Th','Fr','Sa','Su')
count = 0
for i in cal_mth:
    print(i, end = " ")
    count = count +1
    if(count%7 == 0):
        print('')
```

### Step 07 : Publishing Year Calendar
* This code publish the Calendar view for the entire year __(selected in Step 02)__
* Codes from previous cells are applied here
* code can be executed for different values of cal_year (both normal and leap years)

```python
print(f'Calender for year : {cal_year}', end = '\n\n')

for i in mth_numdays_dict.keys():
  which_month = i
  mth = mth_numdays_dict[which_month]
  cal_mth = []

  for i in range(0,mth[1]-1):
    cal_mth = cal_mth + ['  ']
  for i in range(1,mth[0]+1):
    x = str(i)
    if len(x) == 1:
      x = '0'+x
    cal_mth = cal_mth + [x]

  print('='*20)
  print(which_month)
  print('='*20)
  print('Mo','Tu','We','Th','Fr','Sa','Su')
  count = 0
  for i in cal_mth:
    print(i, end = " ")
    count = count +1
    if(count%7 == 0):
        print('')
  print('\n\n')
```
